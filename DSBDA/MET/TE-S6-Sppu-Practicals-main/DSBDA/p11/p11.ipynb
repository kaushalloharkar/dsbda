```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import  GaussianNB
from sklearn.metrics import confusion_matrix,precision_score,r2_score,recall_score,accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Windows 10\OneDrive\Desktop\COLLEGE\TE Sppu Practicals\DSBDA\p11\Iris.csv")
df
```

```python
x=df.drop("Species",axis=1)
```

```python
y=df["Species"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model=GaussianNB()
```

```python
model.fit(xtrain,ytrain)
```

```python
y_pred=model.predict(xtest)
y_pred
```

```python
acc=accuracy_score(ytest,y_pred)
```

```python
#pre=precision_score(ytest,y_pred)
```

```python
err=1-acc
```

```python
print(confusion_matrix.__doc__)
```

```python
confusion_matrix(ytest, y_pred)
```
