```python
data=({"STUDY HOURS":[29,15,33,48,39],"PASS/FAIL":[0,0,1,1,1]})
```

```python
df=pd.DataFrame(data)
df
```

```python
model.fit(x,y)
```

```python
y_pred=model.predict([[25],[42]])
```
